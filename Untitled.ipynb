```python
import os
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# -------- 1.  nc  --------
target_name = "200301_202006-C3S-L3_GHG-PRODUCTS-OBS4MIPS-MERGED-v4.3.nc"
root = r"C:\Users\Lenovo"

file_path = None
for dirpath, dirnames, filenames in os.walk(root):
    if target_name in filenames:
        file_path = os.path.join(dirpath, target_name)
        break

print("", file_path)

# -------- 2.  --------
ds = xr.open_dataset(file_path)

#  ch4 ds 
var = ds["xch4"]

# -------- 3. --------
weights = np.cos(np.deg2rad(ds["lat"]))
global_mean = var.weighted(weights).mean(("lat", "lon"))

# -------- 4.  --------
plt.figure(figsize=(12,5))
global_mean.plot(color="blue")
plt.title("Global Mean Methane (20032020)")
plt.xlabel("Time")
plt.ylabel("CH4")
plt.grid(True)
plt.show()

# -------- 5.  --------
print("\nDescription:")
print("Global mean methane shows a steady increasing trend from 2003 to 2020,")
print("with noticeable acceleration around 20142016 and reaching")
print("its highest values by mid-2020.")
```

```python
import os
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# -------- 1.  nc  --------
target_name = "200301_202006-C3S-L3_GHG-PRODUCTS-OBS4MIPS-MERGED-v4.3.nc"
root = r"C:\Users\Lenovo"

file_path = None
for dirpath, dirnames, filenames in os.walk(root):
    if target_name in filenames:
        file_path = os.path.join(dirpath, target_name)
        break

print("", file_path)

# -------- 2.  --------
ds = xr.open_dataset(file_path)
var = ds["xch4"]     #  ds 

# -------- 3.  (15S, 150W) --------
pt = var.sel(lat=-15, lon= -150, method="nearest")

# -------- 4.  --------
clim = pt.groupby("time.month").mean("time")
deseasonalized = pt - clim.sel(month=pt["time.month"])

# -------- 5.  --------
plt.figure(figsize=(12,5))
deseasonalized.plot(color="darkred")
plt.title("Deseasonalized Methane at (15S, 150W), 20032020")
plt.xlabel("Time")
plt.ylabel("CH4 anomaly")
plt.grid(True)
plt.show()

# -------- 6.  --------
print("\nDescription:")
print("The deseasonalized methane time series at (15S, 150W) shows a clear upward trend,")
print("with interannual variability but consistent long-term increase from 2003 to 2020.")
print("Seasonal effects have been removed, so the curve highlights underlying methane growth.")
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

# 
ds = xr.open_dataset('NOAA_NCDC_ERSST_v3b_SST.nc')

# Nio 3.4
nino34_avg = ds.sst.sel(lat=slice(-5, 5), lon=slice(190, 240)).mean(dim=['lat', 'lon'])

# 
climatology = nino34_avg.groupby('time.month').mean()
anomaly = nino34_avg.groupby('time.month') - climatology

# Nio 3.4
plt.figure(figsize=(12, 6))
anomaly.plot(color='red', linewidth=1)
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.axhline(y=0.5, color='orange', linestyle='--', alpha=0.7, label='El Nio Threshold')
plt.axhline(y=-0.5, color='blue', linestyle='--', alpha=0.7, label='La Nia Threshold')
plt.fill_between(anomaly.time, anomaly, where=anomaly>=0, color='red', alpha=0.3)
plt.fill_between(anomaly.time, anomaly, where=anomaly<0, color='blue', alpha=0.3)
plt.title('Nio 3.4 Index (Sea Surface Temperature Anomaly)')
plt.ylabel('SST Anomaly (C)')
plt.xlabel('Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import os

def verify_merged_data():
    # --- 1.  ---
    file_path = r'C:\Users\Lenovo\MERRA2_400.instU_2d_gas_Nx.201801_202510.nc'
    
    if not os.path.exists(file_path):
        print(f" {file_path}")
        print("")
        return

    print(f": {file_path} ...")
    try:
        ds = xr.open_dataset(file_path)
    except Exception as e:
        print(f": {e}")
        return

    # --- 2.  (Time Dimension) ---
    times = ds['time'].values
    time_len = len(times)
    
    start_time = pd.to_datetime(times[0])
    end_time = pd.to_datetime(times[-1])
    
    print("\n" + "="*40)
    print(" >>>  <<<")
    print("="*40)
    print(f" (Count): {time_len}")
    print(f" (Start): {start_time.strftime('%Y-%m-%d')}")
    print(f" (End)  : {end_time.strftime('%Y-%m-%d')}")
    
    #  (201801 - 202510)
    # 
    expected_months = (end_time.year - start_time.year) * 12 + (end_time.month - start_time.month) + 1
    
    print("-" * 40)
    if time_len == expected_months:
        print(f"  ({time_len}) ")
    else:
        print(f"  ({time_len})  ({expected_months}) ")
        print("   ")

    # --- 3.  (Visual Check) ---
    print("\n" + "="*40)
    print(" >>>  <<<")
    print("="*40)
    
    #  CO, SO2 
    var_name = list(ds.data_vars)[0] 
    print(f" '{var_name}' ...")
    print("")
    print("0")

    # 
    #  mean 
    ds[var_name].mean(dim=['lat', 'lon']).plot(figsize=(12, 6), marker='o', markersize=4)
    
    plt.title(f"Check Merged Data: {var_name} Global Mean ({start_time.year}-{end_time.year})")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    ds.close()

if __name__ == '__main__':
    verify_merged_data()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import os

def process_and_visualize():
    # --- 1.  ---
    file_path = r'C:\Users\Lenovo\MERRA2_400.instU_2d_gas_Nx.201801_202510.nc'
    
    if not os.path.exists(file_path):
        print(f" {file_path}")
        return

    print("...")
    ds = xr.open_dataset(file_path)
    
    #  ( 'COCL', 'SO2' )
    # : var_name = 'COCL'
    var_name = list(ds.data_vars)[0]
    data = ds[var_name]
    print(f": {var_name}")

    # --- 2.  ---
    print("...")
    
    # A.  (NaN)
    #  ()
    data_clean = data.interpolate_na(dim='time', method='linear')
    
    # B.  (Outliers)
    #  99.5%  0 
    # 
    q_low = data_clean.quantile(0.005) #  0.5%
    q_high = data_clean.quantile(0.995) #  99.5%
    
    #  NaN NaN ()
    #  (Clip)
    data_clean = data_clean.clip(min=q_low, max=q_high)

    print("")

    # --- 3.  ( vs ) ---
    print("...")
    plt.figure(figsize=(12, 5))
    
    # 
    ts_raw = data.mean(dim=['lat', 'lon'])
    ts_clean = data_clean.mean(dim=['lat', 'lon'])
    
    ts_raw.plot(label='Original (Raw)', color='lightgray', linestyle='--')
    ts_clean.plot(label='Processed (Cleaned)', color='blue', linewidth=1.5)
    
    plt.title(f'Global Mean Time Series: {var_name} (2018-2025)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- 4.  5  ---
    print(" 5 ...")
    
    fig, axes = plt.subplots(3, 2, figsize=(16, 15))
    #  axes  5 
    ax_list = axes.flatten()

    #  1:  (Time Mean) - 
    mean_map = data_clean.mean(dim='time')
    mean_map.plot(ax=ax_list[0], cmap='viridis', cbar_kwargs={'label': 'Mean Conc.'})
    ax_list[0].set_title('(1) Time Mean (Climatology)')

    #  2: / (Standard Deviation) - 
    std_map = data_clean.std(dim='time')
    std_map.plot(ax=ax_list[1], cmap='plasma', cbar_kwargs={'label': 'Std Dev'})
    ax_list[1].set_title('(2) Variability (Standard Deviation)')

    #  3:  (First Time Step)
    data_clean.isel(time=0).plot(ax=ax_list[2], cmap='magma')
    ax_list[2].set_title(f'(3) Start Date Distribution ({str(data_clean.time[0].values)[:10]})')

    #  4:  (Last Time Step)
    data_clean.isel(time=-1).plot(ax=ax_list[3], cmap='magma')
    ax_list[3].set_title(f'(4) End Date Distribution ({str(data_clean.time[-1].values)[:10]})')

    #  5:  Hovmller  ( vs ) - 
    # 
    zonal_mean = data_clean.mean(dim='lon')
    zonal_mean.plot(ax=ax_list[4], x='time', y='lat', cmap='jet')
    ax_list[4].set_title('(5) Zonal Mean Hovmller (Lat vs Time)')

    #  6 
    fig.delaxes(ax_list[5])

    plt.tight_layout()
    plt.show()
    
    print("")

if __name__ == '__main__':
    process_and_visualize()
```
